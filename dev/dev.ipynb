```python
def visualise(img, axis, slice_index, figsize=(24, 8)):
    plt.figure(figsize=figsize)
    plt.subplot(131)

    if axis == 0:
        plt.imshow(img[slice_index, :, :])#, cmap='gray')
        view = 'Left-right view'
    elif axis == 1:
        plt.imshow(img[:, slice_index, :])#, cmap='gray')
        view = 'Anterior-posterior view'
    elif axis == 2:
        plt.imshow(img[:, :, slice_index])#, cmap='gray')
        view = 'Superior-inferior view'
    else:
        raise ValueError('Not a valid axis.')

    plt.title(view)
```

```python
def visualise_patient(img, mask, axis, slice_index, figsize=(24, 8)):
    if axis == 0:
        view = 'Left-right view'
        img_rot = Image.fromarray(img[slice_index, :, :]).transpose(Image.ROTATE_90) 
        mask_rot = Image.fromarray(mask[slice_index, :, :]).transpose(Image.ROTATE_90) 
    elif axis == 1:
        view = 'Anterior-posterior view'
        img_rot = Image.fromarray(img[:, slice_index, :]).transpose(Image.ROTATE_90) 
        mask_rot = Image.fromarray(mask[:, slice_index, :]).transpose(Image.ROTATE_90) 
    elif axis == 2:
        view = 'Superior-inferior view'
        img_rot = Image.fromarray(img[:, :, slice_index]).transpose(Image.ROTATE_90) 
        mask_rot = Image.fromarray(mask[:, :, slice_index]).transpose(Image.ROTATE_90) 
    else:
        raise ValueError('Not a valid axis.')

    # Plot the image
    plt.figure(figsize=figsize)
    plt.subplot(131)

    plt.imshow(img_rot, cmap='gray')

    plt.title(f'Without aneurysm mask overlay\n{view}')

    # Overlay the aneurysm mask on the image
    cmap_colors = [(0, 0, 0, 0), (0.5, 0, 0.5)]  # Transparent for background, purple for mask
    custom_cmap = ListedColormap(cmap_colors)
    
    plt.subplot(132)
    plt.imshow(img_rot, cmap='gray')
    plt.imshow(np.ma.masked_where(mask_rot == 0, mask_rot), cmap=custom_cmap, alpha=0.9)

    plt.title(f'With aneurysm mask overlay\n{view}')
    plt.tight_layout()
    plt.show()
```

```python
visualise_patient(brain, result, 2, 94)
```

```python
def save_visualisation(img, mask, axis, slice_index, figsize=(24, 10)):
    plt.ioff()
    
    if axis == 0:
        view = 'Left-right view'
        img_rot = Image.fromarray(img[slice_index, :, :]).transpose(Image.ROTATE_90) 
        mask_rot = Image.fromarray(mask[slice_index, :, :]).transpose(Image.ROTATE_90) 
    elif axis == 1:
        view = 'Anterior-posterior view'
        img_rot = Image.fromarray(img[:, slice_index, :]).transpose(Image.ROTATE_90) 
        mask_rot = Image.fromarray(mask[:, slice_index, :]).transpose(Image.ROTATE_90) 
    elif axis == 2:
        view = 'Superior-inferior view'
        img_rot = Image.fromarray(img[:, :, slice_index]).transpose(Image.ROTATE_90) 
        mask_rot = Image.fromarray(mask[:, :, slice_index]).transpose(Image.ROTATE_90) 
    else:
        raise ValueError('Not a valid axis.')

    # Plot the image
    fig, axes = plt.subplots(1, 2, figsize=figsize)

    axes[0].imshow(img_rot, cmap='gray')
    axes[0].set_title(f'Without aneurysm mask overlay\n{view}')
    axes[0].axis('off')

    # Overlay the aneurysm mask on the image
    cmap_colors = [(0, 0, 0, 0), (0.5, 0, 0.5)]  # Transparent for background, purple for mask
    custom_cmap = ListedColormap(cmap_colors)
    
    axes[1].imshow(img_rot, cmap='gray')
    axes[1].imshow(np.ma.masked_where(mask_rot == 0, mask_rot), cmap=custom_cmap, alpha=0.9)
    axes[1].set_title(f'With aneurysm mask overlay\n{view}')
    axes[1].axis('off')

    plt.tight_layout()
    
    plt.savefig('./test_save.png')
```

```python
save_visualisation(brain, result, 2, 94)
```
